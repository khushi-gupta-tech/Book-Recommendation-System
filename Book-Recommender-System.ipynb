
# Popular Based Recommender System

```python
ratings_with_books = ratings.merge(books,on= "ISBN" )
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
num_rating = ratings_with_books.groupby('Title').count()['Rating'].reset_index()
num_rating.rename(columns = {
    "Rating" : "num_rating"},
    inplace = True)
num_rating.head()
```

```python
avg_rating = ratings_with_books.groupby('Title')['Rating'].mean().reset_index()

avg_rating.rename(columns = {
    "Rating" : "avg_rating"},
    inplace = True)
avg_rating.head()
```

```python
popular_df = num_rating.merge(avg_rating,on='Title')
popular_df.head()
```

```python
popular_df = popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
final_df = popular_df.merge(books,on='Title').drop_duplicates('Title')[['Title','Author','num_rating','avg_rating','img_url']]
```

```python
final_df
```

```python
final_df['img_url'][0]
```

# Collaborative Filtering Based Recommender System

```python
x = ratings_with_books.groupby('User_id').count()['Rating'] > 200
edu_users = x[x].index
edu_users
```

```python
filtered_rating = ratings_with_books[ratings_with_books['User_id'].isin(edu_users)]
```

```python
y = filtered_rating.groupby('Title').count()['Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Title',columns='User_id',values='Rating')
```

```python
pt.fillna(0,inplace= True)
```

```python
pt
```

# Modelling

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(pt)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distance,suggestion = model.kneighbors(pt.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
def recommend(book_name):
    # index fetch
    book_id = np.where(pt.index==book_name)[0][0]
    distance,suggestion = model.kneighbors(pt.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    data = []
    for i in suggestion:
        item = []
        temp_df = books[books['Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Title')['Title'].values))
        item.extend(list(temp_df.drop_duplicates('Title')['Author'].values))
        item.extend(list(temp_df.drop_duplicates('Title')['img_url'].values))

        data.append(item)

    return data
```
